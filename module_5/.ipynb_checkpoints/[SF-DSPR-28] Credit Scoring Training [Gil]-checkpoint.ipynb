```python
def intitial_eda_checks(df, missing_percent):
    '''
    Функция принимает на вход датафрейм, а также заданный порог % пустых значений, который хотим обработать. 
    На выход выводит на экран информацию о сумме пустых значений для всех колонок, а также проце
    '''
    if df.isnull().sum().sum() > 0:
        mask_total = df.isnull().sum().sort_values(ascending=False)
        total = mask_total[mask_total > 0]

        mask_percent = df.isnull().mean().sort_values(ascending=False)
        percent = mask_percent[mask_percent > 0]

        series = mask_percent[mask_percent > missing_percent]
        columns = series.index.to_list()

        missing_data = pd.DataFrame(pd.concat(
            [total, round(percent*100, 2)], axis=1, keys=['Количество', '%']))
        print('Сумма и процент значений NaN:\n')
        display(missing_data)
    else:
        print('NaN значения не найдены.')
```

```python
# Запускаем функцию вывода всех пустых значений
intitial_eda_checks(train, 0)
```

## Визуализация
Посмотрим на распределение числовых данных

```python
for col in num_cols:
    plt.figure()
    sns.distplot(train[col], rug=False)
    plt.title(col)
    plt.show()
```

После построения гистограмм стало очевидно, что распределения всех числовых переменных имеют тяжёлый правый хвост. 

…Насколько я помню, для того чтобы сделать распределение данных переменных более нормальным, можно работать с логарифмированными величинами этих переменных. Тогда я смогу избежать чувствительности к сильным отклонениям в суммах у линейных моделей.


```python
for i in num_cols:
    plt.figure()
    sns.distplot(train[i][train[i] > 0].dropna(), kde = False, rug=False)
    plt.title(i)
    plt.show()
```

```python
def get_boxplot(column):
    """
    Функция для отрисовки коробочной диаграммы для нечисловых величин.
    На вход получаем список колонок для отрисовки. 
    Отрисовываем относительно целевой переменной Rating.
    """
    fig, ax = plt.subplots(figsize=(14, 4))
    sns.boxplot(x='default', y=column,
                data=train,
                ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot для ' + column)
    plt.show()
```

```python
for col in num_cols:
    get_boxplot(col)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 8))
plt.subplots_adjust(wspace=0.5)
axes = axes.flatten()

for i in range(len(num_cols)):
    sns.boxplot(x='default', y=num_cols[i], data=train,
                orient='v', ax=axes[i], showfliers=False)
```
