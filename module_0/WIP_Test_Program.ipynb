```python
#version_1
import numpy as np
number = np.random.randint(1,101)    # загадали число

search_list = [x for x in range(1, 101, 1)]
predict = search_list[(len(search_list)//2)-1]
count = 0
print ("Загадано число от 1 до 100.")
print ('')

while len(search_list)>0:
    count += 1 # плюсуем попытку
    print (f"Ваше число {predict}? Это попытка номер {count}")
    index = 0
    if number == predict: 
        print('Да, угадали!')
        break    # выход из цикла, если угадали
    elif number > predict: 
        for item in search_list:
            if item == predict: break
            else: index +=1
        search_list = search_list[index+1:]
        print(f"Нет, мое число больше {predict}")
    elif number < predict: 
        for item in search_list:
            if item == predict: break
            else: index +=1
        search_list = search_list[:index]
        print(f"Нет, мое число меньше {predict}")
    predict = search_list[(len(search_list)//2-1)]
    print(search_list, '', number) 
    print('')

print (f"Вы угадали число {number} за {count} попыток.")
```

```python
#version_2
import numpy as np
number = np.random.randint(1,101)    # загадали число

N = 100
# заполнение массива 
search_list = []
search_list = [x for x in range(1, N+1, 1)]
low_index = 0
high_index = N-1
mid_index = (low_index + high_index)//2
count = 0
    
print ("Загадано число от 1 до 100.")
print ('')

# как только нижний индекс станет больше на 1 верхнего 
# или верхний на 1 меньше нижнего цикл остановится
while low_index <= high_index:
    
    count += 1 # плюсуем попытку
    
    print (f"Ваше число {search_list[mid_index]}? Это попытка номер {count}")

    if number < search_list[mid_index]:
        # то верхняя граница сдвигается к середине (но на 1 до нее, т. к. середина была уже проверена)
        high_index = mid_index - 1
    # Если искомое число больше числа с индексом середины,
    elif number > search_list[mid_index]:
        # то нижняя граница сдвигается за середину
        low_index = mid_index + 1
    # Все остальные случаи возникают, когда искомое число 
    else:
        print("ID =", mid_index, "number = ", search_list[mid_index])
        # прерывание цикла
        break
    
    # находится индекс середины массива или отрезка массива
    mid_index = ((low_index + high_index)//2-1)
    
    print(low_index, high_index, search_list[mid_index], number) 
    print('')

print (f"Вы угадали число {number} за {count} попыток.")
```
